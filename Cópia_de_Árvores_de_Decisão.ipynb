```
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics

import pandas as pd
```

```
def run_decision_tree(caminho : str) -> None:
  url = caminho
  dataset = pd.read_csv(url, header=None)

  columns = len(dataset.columns)

  y = dataset[0]
  X = dataset.loc[:,1:columns-1]
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None, stratify=y) # 80% treino e 20% teste
  model = tree.DecisionTreeClassifier(criterion="entropy")
  model = model.fit(X_train, y_train)

  result = model.predict(X_test)
  acc = metrics.accuracy_score(result, y_test)
  show = round(acc * 100)
  print("{}%".format(show))

  print(list(result))
  print(list(y_test))

  tree.plot_tree(model,filled=True)


```

```
parkinson_caminho = "https://raw.githubusercontent.com/tmoura/machinelearning/master/datasets/parkinsons.data"
run_decision_tree(caminho = parkinson_caminho)
```

```
sonar_caminho = "https://raw.githubusercontent.com/tmoura/machinelearning/master/datasets/sonar.data"
run_decision_tree(caminho = sonar_caminho)
```
